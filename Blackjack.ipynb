```python
import random
```

```python
#Variables

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}
```

```python
class Card:
    
    def __init__(self,rank,suit):
        
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
    
    def __str__(self):
        
        return self.rank + ' of ' + self.suit
```

```python
class Deck:
    

    
    
    def __init__(self):
        
        self.all_cards = []
        
        for s in suits:
            for r in ranks:
                self.all_cards.append(Card(r,s))
    
    def shuffle(self):
        
        import random
        
        random.shuffle(self.all_cards)

    def deal_one(self):
        
        return self.all_cards.pop()
        
     
  
```

```python
# Introduction to the game

def intro():
    
    name = input("Welcome to the blackjack! Please enter your name: ")
    
    return name
```

```python
#Ask player for their bet

def ask_bet(player_money = 100):
    
    bet = 10**20
    
    while bet >= player_money:

        bet = int(input(f"Please provide the bet. Remember it can not be higher than your current wallet state = {player_money}!: \n"))
    return bet
```

```python
#game loop
def game():
    
    deck_one = Deck() #Deck to play with
    deck_one.shuffle() #shuffle the deck
    
    # some variables 
    player_money = 100
    player_hand_value = 0
    pc_hand_value = 0
    
    #Ask player for the bet.
    bet = ask_bet()
    player_money -= bet

    # Player and pc hands:
    player_hand = []
    pc_hand = []

    #now deal 2 cards to each player:
    for player in range(2):

        #for player:
        dealt_card = deck_one.deal_one()
        player_hand.append(dealt_card)
        player_hand_value += dealt_card.value

        #for croupier
        dealt_card = deck_one.deal_one()
        pc_hand.append(dealt_card)
        pc_hand_value += dealt_card.value

    print(f"The dealer last card is {dealt_card}\n")
    print("Your hand: \n")
    for card in player_hand:
        print(card)
    print(f"Your hand value is {player_hand_value}\n")

    #Now player makes a decision whether He/She wants to play or pass.

    player_satisfied = False
    dealer_satisfied = False
    while player_satisfied == False or dealer_satisfied == False: 
        move = 'wrong'
        if pc_hand_value < 17: #Check if the dealer exceeded 16, if not, hit.
            dealt_card = deck_one.deal_one()
            pc_hand.append(dealt_card)
            pc_hand_value += dealt_card.value
        else:
            dealer_satisfied = True
        
        if player_hand_value > 20: # Check whether player busted or not. If he did, break the loop.
            player_satisfied = True
            print("You have busted 21! \n")
        else:
            move = input("Would you like to hit(h) or stand(s)?\n")
            
        if move == "h": # Player hits. Add another card to his hand. 
            dealt_card = deck_one.deal_one()
            player_hand.append(dealt_card)
            player_hand_value += dealt_card.value  
            print("Your hand: \n")
            for card in player_hand:
                print(card)
            print(f"Your hand value is {player_hand_value}\n")

        elif move == 's':
            player_satisfied = True

    #Lets check the winner and give away the bet.

    if abs(player_hand_value - 21) < abs(pc_hand_value - 21):
        print(f"You have won! good job. Your hand value was {player_hand_value} and pc hand was {pc_hand_value}. ")
        winner = "Player one"
        player_money += bet*2
        print(f"You have now total of {player_money}!")
    
    elif abs(player_hand_value - 21) > abs(pc_hand_value - 21):
        print(f"This time AI have won. Your hand value was {player_hand_value} and pc hand was {pc_hand_value}. ")
        winner = "PC"
        print(f"You have now total of {player_money}!")
    
    else:
        print (f"It is a tie! Your hand value was {player_hand_value} and pc hand was {pc_hand_value}.")
        winner = 'None'
        player_money +=bet
        print(f"You have now total of {player_money}!")
        
    return winner
```

```python
def blackjack():
    
    name = intro()
    
    winner = game()
    
    print(f"\nThe winner is {winner}")
```
